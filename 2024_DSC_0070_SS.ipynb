```python
mean_income = df['Income-wages'].mean()
median_income = df['Income-wages'].median()
print(f"Mean yearly income: {mean_income:,.2f}$")
print(f"Median yearly income: {median_income:,.2f}$")
```

```python
#b)
std_dev = df['Income-wages'].std() 
variance = df['Income-wages'].var()
print(f"Standard Deviation of yearly income: {std_dev:,.2f}$")
print(f"Variance of yearly income: {variance:,.2f}$")
#calculate the standard deviation and varience
```

```python
#c)
no_variability_std_dev = 0 
no_variability_IQR = 0
print(f"Standard deviation with no variability: {no_variability_std_dev}")
print(f"IQR with no variability: {no_variability_IQR}")
#If there is no variability then both standard deviation and IQR are both 0
```

```python
#d)
no_variability_mean = df['Income-wages'].mean()
no_variability_median = df['Income-wages'].median()
no_variability_min = df['Income-wages'].min()
no_variability_max = df['Income-wages'].max()
no_variability_Q1 = df['Income-wages'].quantile(0.25)
no_variability_Q3 = df['Income-wages'].quantile(0.75)
print(f"Mean income: ${no_variability_mean:,.2f}")
print(f"Median income: ${no_variability_median:,.2f}")
print(f"Q1: ${no_variability_Q1:,.2f}")
print(f"Q3: ${no_variability_Q3:,.2f}")
print(f"Minimum income: ${no_variability_min:,.2f}")
print(f"Maximum income: ${no_variability_max:,.2f}")
```

```python
#e)
plt.figure(figsize=(8, 5)) 
sns.boxplot(data=df, x ='Education', y='Income-wages',color="green")
plt.title('Income Distribution Based On Education Level')
plt.xticks(rotation=90)
plt.show()
#Boxplot visualization of income distribution based on education level
```

```python
plt.figure(figsize=(10, 6)) 
sns.boxplot(data=df, x='Region-division_recode', y='Income-wages',color="pink")
plt.title('Income Distribution Based On Region')
plt.xticks(rotation=90)
plt.show()
#Boxplot visualization of income distribution based on region
```
